```python
# HeatMapPlot_fromCSVfile
# This code reads an csv file, which contains the information in form of a table (e.g. year, names, data, 5*6),
#  and plots it as a  heatmap plot.
# python version  3.12.2
# 06/21/26

import sys
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

# checking the interpreter in use -optional
print('Interpreter = ')
print(sys.executable) #prints the path of the interpreter you are using: i.e./Users/yuriria/miniconda3/envs/intro_to_python/bin/python
print(sys.version_info)

# 1. read the data from the csv file, extract relevant variables. The data are a table (year, names,data)
filename = '/Users/yuriria/Dropbox/Python_related/Intro_To_Python-master-2/docs/data/salmon_mercury.csv'
data = np.genfromtxt(filename, delimiter = ',', unpack = False, skip_header = True, usecols = [1,2,3,4,5])
names = np.genfromtxt(filename, delimiter = ',', unpack = True, skip_header = True, usecols = [0], dtype = str)
years = np.genfromtxt(filename, delimiter = ',', unpack = False, skip_header = False, dtype = str,)[0][1:]


# 2. Create heatmap plot & labels
fig, ax = plt.subplots()
sns.heatmap(data, linewidth = 0.5, cmap='viridis', annot = True)
ax.set_xlabel("Year")
ax.set_ylabel("Salmon Species")
ax.set_title("Mercury Content  Over Time (ppm)")
ax.set_yticklabels(names)
ax.set_xticklabels(years)
ax.tick_params(axis='y', labelrotation=0)
plt.show()

```
