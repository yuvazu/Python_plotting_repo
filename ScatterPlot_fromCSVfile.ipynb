```python
# Scatter Plot_fromCSVfile

# This code reads an csv file and make a Scatter plot for two of the measured categories i.e. length vs weight
# It assumes the first row in the "csv" file contains the category names
# and the rest of the rows are the values. The values per category have the "same" lengths. 
# python version  3.12.2
# 06/21/26

# checking the interpreter in use -optional
import sys
import numpy as np 
import matplotlib.pyplot as plt

print('Interpreter = ')
print(sys.executable) #prints the path of the interpreter you are using: i.e./Users/yuriria/miniconda3/envs/intro_to_python/bin/python
print(sys.version_info)

# 1. read the data from the csv file
# modify filename accordingly
filename = '/Users/yuriria/Dropbox/Python_related/Intro_To_Python-master-2/docs/data/salmon_catch_data.csv' 
id, name, length, weight = np.genfromtxt(filename, delimiter = ',', unpack = True, skip_header = True, dtype=None, encoding = None)
#print(name)


# 2. Append the values to the corresponding variable
chinook_length = []
chinook_weight = []

for i in range(len(length)):
    if weight[i] > 0:
        chinook_weight.append(weight[i])
        chinook_length.append(length[i])

print(chinook_length) 
print(chinook_weight)

fig, ax = plt.subplots()
ax.scatter(chinook_length, chinook_weight, c = 'lightsalmon', edgecolor = 'salmon', s = 50, marker = 'D')
ax.set_title('Chinook Salmon Length vs Weight')
ax.set_xlabel('Length (mm)')
ax.set_ylabel('Weight (g)')
plt.show()
```
